```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(10)
y = x ** 2 
plt.figure()
plt.plot(x, y, 'b')
plt.xlabel("Time (s)")
plt.ylabel("Velocity (m/s)")
plt.title("Velocity vs. Time Graph")
plt.show()
```

Using matplotlib's object-oriented plotting:

```python
fig = plt.figure()
ax = fig.add_axes([80, 40, 0.8, 0.8]) # x, y (bottom-left corner), width, height 
plt.plot(x, y, 'g')
ax.set_xlabel("Time (s)")
ax.set_ylabel("Velocity (m/s)")
ax.set_title("Velocity vs. Time Graph")
```

Can add multiple subplots to one figure:

```python
fig = plt.figure()
ax1 = fig.add_axes([0, 0, 0.8, 0.8])
ax1.plot(x, y, 'g')
ax2 = fig.add_axes([0.1, 0.4, 0.3, 0.3])
ax2.plot(y, x, 'b')
```

Instead of add_axes, use subplots if you don't care about exact positioning:

```python
fig, ax = plt.subplots()
ax.plot(x, y)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
for ax in axes:
    ax.plot(x, y)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
fig.tight_layout() # Prevents the right y-label from overlapping with the left graph
```

Setting things like figsize, dpi, etc:

```python
fig = plt.figure(figsize=(8, 4), dpi=100)
fig, ax = plt.subplots(figsize=(6, 4), dpi=60)
ax.plot(x, y)
```

Multiple graphs in one subplot and adding legends:

```python
fig, ax = plt.subplots()
ax.plot(x, y, label='Velocity')
ax.plot(x, 2 * y ** 0.5, label='Acceleration')
ax.legend()
# ax.legend(loc=0) loc=0, 1, 2, 3, 4... => 0: placed in best location, 1: Q1 (upper-right), 2: Q2, 3: Q3, 4: Q4
```

Changing style using rcParams, setting back to default, and LaTeX:

```python
import matplotlib
matplotlib.rcParams.update({'font.size': 18, 'font.family': 'serif'})
fig, ax = plt.subplots()
ax.plot(x, y, label=r"$\overrightarrow{v} = t^2$")
ax.plot(x, 2 * y ** 0.5, label=r"$\overrightarrow{a}=2t$")
ax.legend()
matplotlib.rcParams.update({'font.size': 12, 'font.family': 'sans'}) # Back to default
```

Line styles (linewidth, color, dotted, etc.): 

```python
fig, ax = plt.subplots()
ax.plot(x, x, 'r.-')        # Red, dots, and line
ax.plot(x, x + 1, 'b--')    # Blue, dashed line
ax.plot(x, x + 2, 'g*--')   # Green, starts, dashed line
ax.plot(x, x + 3, 'y.')     # Yellow, dots
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(x, x, color='violet', linewidth=0.5)
ax.plot(x, x + 1, color='violet', lw=1)           # linewidth = lw
ax.plot(x, x + 2, color='violet', lw=2)

# linestyle: -, --, -., :, steps
ax.plot(x, x + 3, color='indigo', linestyle='-')
ax.plot(x, x + 4, color='indigo', ls='--')        # linestyle = ls
ax.plot(x, x + 5, color='indigo', ls='-.')
ax.plot(x, x + 6, color='blue', ls=':')
ax.plot(x, x + 7, color='blue', ls='steps')

# marker: +, o, *, s, ',', '.', 1, 2, 3, 4, ...
ax.plot(x, x + 9, color='green', marker='+')
ax.plot(x, x + 10, color='green', marker='o')
ax.plot(x, x + 11, color='green', marker='*')
ax.plot(x, x + 12, color='green', marker='s')
ax.plot(x, x + 13, color='orange', marker=',', markersize=2, lw=0.25)
ax.plot(x, x + 14, color='orange', marker='.', markersize=4, lw=0.25)
ax.plot(x, x + 15, color='orange', marker='1', markersize=8, lw=0.25)
ax.plot(x, x + 16, color='orange', marker='2', markersize=8, lw=0.25)
ax.plot(x, x + 16, color='red', marker='o', markersize=8, markerfacecolor='yellow')
ax.plot(x, x + 17, color='red', marker='s', markersize=8, markerfacecolor="yellow", 
        markeredgewidth=2, markeredgecolor="purple")
```

Custom axis ranges:

```python
fig, axes = plt.subplots(1, 2, figsize=(8, 4))
axes[0].plot(x, x ** 1, x, x ** 2)
axes[0].set_title("Default")
axes[1].plot(x, x ** 1, x, x ** 2)
axes[1].set_xlim((0, 6))
axes[1].set_ylim((0, 30))
axes[1].set_title("Custom")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(8, 4))
axes[0].plot(x, 2 ** x)
axes[0].set_title("Normal Scale")
axes[1].plot(x, 2 ** x)
axes[1].set_yscale('log')
axes[1].set_title("Logarithmic Scale")
```

```python
fig, ax = plt.subplots()
ax.plot(x, 2 ** x)
ax.set_yticks([1, 32, 64, 128, 256])
ax.set_ylim([0, 256])
ax.set_yticklabels(['2^0', '2^5', '2^6', '2^7', '2^8'])
ax.set_xticks([1, 3, 5, 7, 9])
ax.set_xticklabels(['one', 'three', 'five', 'seven', 'nine'])
```

Grids:

```python
fig, axes = plt.subplots(1, 2)
axes[0].plot(x, x ** 2)
axes[0].grid(True)
axes[1].plot(x, x **2)
axes[1].grid(color='r', ls='dashdot', lw=1) # âsolidâ, âdashedâ, âdashdotâ, âdottedâ
```

Axis spines:

```python
fig, ax = plt.subplots(figsize=(6, 2))
ax.spines['bottom'].set_color('red')
ax.spines['left'].set_color('orange')
ax.spines['top'].set_color('green')
ax.spines['right'].set_color('none')
ax.spines['left'].set_linewidth(2)
ax.xaxis.tick_top()
```

Twin axes: 

```python
fig, ax1 = plt.subplots()
ax1.plot(x, x ** 2, color='blue')
ax1.set_ylabel("Position (m)", color='blue')
for label in ax1.get_yticklabels():
    label.set_color('blue')

ax2 = ax1.twinx()
ax2.plot(x, 2 * x, color='green')
ax2.set_ylabel("Velocity (m/s)", color='green')
for label in ax2.get_yticklabels():
    label.set_color('green')
```

Setting (0, 0) position to the center of the graph: 

```python
fig, ax = plt.subplots()
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position(('data', 0)) # Place spine at specified data coordinate
ax.spines['left'].set_position(('data', 0))
# set_position: takes a tuple in the form ('data' | 'outward | 'axes', value)
x1 = np.linspace(-2, 2, 20)
ax.plot(x1, x1 ** 3)
```

Other types of plots: 

```python
fig, axes = plt.subplots(1, 4, figsize=(12, 3))
axes[0].scatter(x1, x1 + 0.5 * np.random.randn(len(x1))) 
# randn returns array of same length as x, with random values from the standard normal distribution.
axes[0].set_title("Scatter Plot")
axes[1].step(x, x ** 2)
axes[1].set_title("Step Plot")
axes[2].bar(x, x ** 2)
axes[2].set_title("Bar Plot")
axes[3].fill_between(x, x ** 2, x ** 2.5)
axes[3].set_title("Fill Between Plot")
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 0.8, 0.8], polar=True)
x2 = np.linspace(0, 2 * np.pi, 100)
ax.plot(x2, 4 + np.cos(6 * x2))
ax.plot(x2, x2)
ax.plot(x2, 6 * np.sin(3 * x2))
ax.set_ylim(0, 8)
ax.set_xticklabels(['0', r'$\pi/4$', r'$\pi/2$', r'$3\pi/4$', r'$\pi$', r'$5\pi/4$', r'$3\pi/2$', r'$7\pi/4$', r'$\pi/4$'])
```

```python
fig, axes = plt.subplots(1, 2, figsize=(8, 4))
x3 = np.random.randn(10000)
axes[0].hist(x3, bins=30, edgecolor='black')
axes[0].set_title("Normal Histogram")
axes[1].hist(x3, cumulative=True, bins=50, edgecolor='white')
axes[1].set_xlim(min(x3), max(x3))
axes[1].set_title("Cumulative Histogram")
```

Adding text: 

```python
fig, ax = plt.subplots()
ax.plot(x, x ** 2, x, x ** 3)
ax.text(7, 100, r'$y = x^2$')
ax.text(6, 400, r'$y = x^3$')
```

Customizing grid layout: 

```python
fig = plt.figure()
ax1 = fig.add_subplot(231)
ax1.plot(x, x)
ax2 = fig.add_subplot(2, 3, 2)
ax2.plot(x, x ** 2)
ax3 = fig.add_subplot(236)
ax3.plot(x, x ** 3)
fig.tight_layout()
```

```python
fig, axes = plt.subplots(2, 3)
for i in range(len(axes)):
    for j in range(len(axes[i])):
        axes[i,j].plot(x, x ** (j + i))
fig.tight_layout()
```

```python
fig = plt.figure()
ax1 = plt.subplot2grid((3, 3), (0, 0), colspan=2)
ax1.plot(x, x)
ax2 = plt.subplot2grid((3, 3), (0, 2), rowspan=2)
ax2.plot(x, x ** 2)
ax3 = plt.subplot2grid((3, 3), (1, 0), rowspan=2, colspan=2)
ax3.plot(x, x ** 3)
ax4 = plt.subplot2grid((3, 3), (2, 2))
ax4.plot(x, x ** 4)
fig.tight_layout()
```

Colormaps and contours:

```python
def flux_qubit_potential(phi_m, phi_p):
    return 2.7 - 2 * np.cos(phi_m) * np.cos(phi_p) - 0.7 * np.cos(np.pi - 2 * phi_p)

phi_m = np.linspace(0, 2 * np.pi, 100)
phi_p = np.linspace(0, 2 * np.pi, 100)
X, Y = np.meshgrid(phi_m, phi_p)
Z = flux_qubit_potential(X, Y).T
```

```python
fig, ax = plt.subplots()
p = ax.pcolor(X / (2 * np.pi), Y / (2 * np.pi), Z, cmap=matplotlib.cm.winter, vmin=abs(Z).min(), vmax=abs(Z).max())
cb = fig.colorbar(p, ax=ax)
```

```python
fig, ax = plt.subplots()
im = ax.imshow(Z, cmap=matplotlib.cm.RdBu, vmin=abs(Z).min(), vmax=abs(Z).max(), extent=[0, 1, 0, 1])
im.set_interpolation('bilinear')
cb = fig.colorbar(im, ax=ax)
```

```python
fig, ax = plt.subplots()
cnt = ax.contour(Z, cmap=matplotlib.cm.spring, vmin=abs(Z).min(), vmax=abs(Z).max(), extent=[0, 1, 0, 1])
```

3D plots:

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

```python
fig = plt.figure(figsize=(14, 6))
ax1 = fig.add_subplot(121, projection='3d')
X1, Y1 = np.meshgrid(x1, x1)
Z1 = -np.sin(np.sqrt(X1 ** 2 + Y1 ** 2))
p = ax1.plot_surface(X1, Y1, Z1)

ax2 = fig.add_subplot(122, projection='3d')
p = ax2.plot_surface(X1, Y1, Z1, cmap=matplotlib.cm.RdBu, cstride=1, rstride=1, antialiased=False)
cb = fig.colorbar(p, shrink=0.6)
```

3D wireframe plot:

```python
fig = plt.figure(figsize=(7, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(X1, Y1, Z1, cstride=2, rstride=2)
```

3D plots with contours:

```python
fig = plt.figure(figsize=(7, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X1, Y1, Z1, alpha=0.2)

ax.set_xlim3d([-4, 4])
ax.set_ylim3d([-4, 4])
ax.set_zlim3d([-1.5, 0])

cset = ax.contour(X1, Y1, Z1, zdir='x', offset=-4)
cset = ax.contour(X1, Y1, Z1, zdir='y', offset=4)
cset = ax.contour(X1, Y1, Z1, zdir='z', offset=-1.5)
```

Changing the view angle: 

```python
fig = plt.figure(figsize=(14, 6))
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(X1, Y1, Z1)
ax1.view_init(20, 10) 
# (elevation, azimuth) -> azimuth is like the angle of rotation of the bottom 2D grid. Both in degrees. 
ax2 = fig.add_subplot(122, projection='3d')
ax2.plot_wireframe(X1, Y1, Z1)
ax2.view_init(70, 70)
```
