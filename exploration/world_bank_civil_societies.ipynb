```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('world_bank_civil_society.csv')
df.head(10)
```

# By Country

```python
data = df.groupby('Country Name').mean()
data.head(15)
```

```python
countries = ['Afghanistan', 'Iran, Islamic Rep.', 'Ghana', 'India', 'Russian Federation', 'Nigeria', 'China']
fig, ax = plt.subplots(figsize=(10, 6))
ax.set_title('Average Score For Select Countries')
ax.plot(data.reindex(countries).T)
ax.legend(countries, loc=1)
plt.show()
```

```python
data.describe()
```

```python
print('Best countries in 2018:')
data.sort_values('2018', ascending=False)['2018'].head(10)
```

```python
print('Worst countries in 2018:')
data.sort_values('2018')['2018'].head(10)
```

```python
fig, ax = plt.subplots()
ax.set_title('Overall Trend for Democracy')
ax.plot(data.mean())
ax.set_ylim(5.3, 5.7)
plt.show()
```

```python
indicators = df.groupby('Indicator').mean()
indicators.head()
```
